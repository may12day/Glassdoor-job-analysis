```python
for i in range(10,20):
    if(i==1):
        driver.get("https://www.glassdoor.co.in/Job/india-data-science-jobs-SRCH_IL.0,5_IN115_KO6,18.htm")
    else:
        driver.get("https://www.glassdoor.co.in/Job/india-machine-learning-jobs-SRCH_IL.0,5_IN115_KO6,22_IP"+str(i)+".htm")
    driver.implicitly_wait(4)
    all_jobs = driver.find_elements_by_class_name('jl')
    for job in all_jobs:

        result_html = job.get_attribute('innerHTML')
        soup = BeautifulSoup(result_html,'html.parser')

        try:
            company = soup.find("div",class_="jobHeader").text.replace('\n','')
        except:
            company = 'None'

        tit_len = len(company)

        try:
            location = soup.find(class_="subtle loc").text
        except:
            location = 'None'

        loc_len = location[0]

        try:
            title = soup.find('div', {'class' :'jobContainer'}).text[tit_len:]
            title = title[:title.index(loc_len)]
        except:
            title = 'None'

        try:
             rating = soup.find("span",class_="compactStars").text
        except:
            rating = 'None'

        sum_div = job.find_elements_by_class_name("jobHeader")[0]
        try:
            sum_div.click()
        except:
            close_button = driver.find_elements_by_class_name("ModalStyle__xBtn___34qya")[0]
            close_button.click()
            sum_div.click()

        job_desc = driver.find_element_by_class_name('jobDescriptionContent').text.replace('\n',' ')


        dataframe2 = dataframe2.append({'Title':title,'Location':location,"Company":company,
                                      "Rating":rating,"Description":job_desc},
                                     ignore_index=True)   




```
